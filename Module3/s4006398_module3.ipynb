### **Studio Class (SC) 9, Question 1:** <a id="q9a"></a>

Why are Scenarios and Storyboards part of the User Centred Design process?

Scenarios and storyboards are integral to the User-Centered Design process as they enable designers to empathise with users, analyse user contexts and tasks, iterate on design ideas, communicate and collaborate effectively, and validate design concepts. By creating narratives and visual representations, designers can better understand user needs, explore different design options, gather feedback, and ensure that the final solution meets user expectations.

### **Studio Class (SC) 9, Question 2:** <a id="q9b"></a>

What the key benefits of Scenarios and Storyboards?

Ensures the designs is workable, gives an idea of what acitons must be taken by a user when interacting with the solution, uncovering error conditions, predicting user behavior and preventing rework.

### **Studio Class (SC) 9, Question 3:** <a id="q9c"></a>

What the three important elements for creating good quality scenarios?

When creating good quality scenarios, it is crucial to maintain a user-centered focus, ensuring that the scenarios reflect the user's needs, goals, and motivations. Additionally, scenarios should provide sufficient context and detail, painting a clear picture of the user's environment, actions, and interactions. It is important to create scenarios that are relevant to the design objectives and cover a range of user profiles and situations to capture diverse user needs. By incorporating these elements, designers can develop scenarios that effectively communicate the user's experience and guide the design process.

### **Studio Class (SC) 9, Question 4:** <a id="q9d"></a>

How do storyboards relate to scenarios?

Storyboards and scenarios are interconnected tools used in visual storytelling, where scenarios outline the written narrative, including plot and dialogue, while storyboards visually depict key scenes through hand-drawn or digital illustrations, enabling the creative team to plan and communicate the visual aspects of the story, such as composition and camera angles. Together, they ensure a cohesive and effective storytelling experience by translating written narratives into visual representations.
