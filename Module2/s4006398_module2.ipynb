PRAGMA foreign_keys = OFF;
drop table if exists COMPANY;
drop table if exists PRODUCT;
drop table if exists Stock;
PRAGMA foreign_keys = ON;

CREATE TABLE COMPANY
        (COMPANYID  INTEGER NOT NULL,
         TITLE      VARCHAR(30)        ,
         PRIMARY KEY (COMPANYID)        
        )
        WITHOUT ROWID;

CREATE TABLE PRODUCT
        (PRODUCTID  INTEGER NOT NULL,
         TITLE      VARCHAR(30)        ,
         PRIMARY KEY (PRODUCTID)      
        )
        WITHOUT ROWID;

CREATE TABLE STOCK
        (COMPANYID  INTEGER NOT NULL,
         PRODUCTID  INTEGER NOT NULL,
         QUANTITY   INTEGER NOT NULL,
         PRIMARY KEY (COMPANYID, PRODUCTID),
         FOREIGN KEY(COMPANYID) REFERENCES COMPANY(COMPANYID),
         FOREIGN KEY (PRODUCTID) REFERENCES PRODUCT(PRODUCTID)    
        )
        WITHOUT ROWID;

Explain how the returned rows of a query can be ordered, and provide an example.

The returned rows of a query can be ordered using the ORDER BY clause in SQL. It allows you to specify the column or columns by which you want to sort the result set in either ascending (ASC) or descending (DESC) order. You can also specify multiple columns to perform sorting based on multiple criteria, providing flexibility in how the rows are ordered. e.g SELECT * FROM employees ORDER BY salary DESC;
