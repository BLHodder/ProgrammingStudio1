### **Studio Class (SC) 13, Question 2:** <a id="q13b"></a>

What is the essential difference between a WHERE clause and the HAVING clause?

The WHERE clause is used to filter rows based on conditions applied to individual columns, narrowing down the result set. The HAVING clause filters aggregated data based on conditions applied to the result of a GROUP BY clause. WHERE operates on individual rows, while HAVING operates on grouped and aggregated data.

### **Studio Class (SC) 13, Question 3:** <a id="q13c"></a>

Explain what a trigger is and why we use them.

Triggers are automated actions associated with database tables that execute when specific events occur, such as data modifications. They are used to enforce business rules, maintain data integrity, and automate actions within a database system, improving functionality and reliability.

### **Studio Class (SC) 13, Question 5:** <a id="q13e"></a>

Explain two advantages of using a view.

Views provide a way to encapsulate complex queries into a single, reusable object. By creating a view, the complexity of the underlying query is abstracted away, allowing users to query the view with a simpler, more intuitive syntax. This simplification reduces the chance of errors and enhances productivity by eliminating the need to repeatedly write and maintain complex queries.

Views enhance security by providing controlled access to sensitive data and allowing selective permissions. They enable hiding columns and filtering rows, ensuring data privacy for different user roles or departments.
